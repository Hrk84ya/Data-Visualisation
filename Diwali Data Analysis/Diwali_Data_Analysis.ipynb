```python
#importing the required modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#reading the dataset
df=pd.read_csv("Diwali Sales Data.csv",encoding='unicode_escape')
#using 'unicode_escape' to avoid encoding error
```

```python
#obtaining the number of rows and columns
df.shape
```

```python
#printing the first 10 rows of the dataset
df.head(10)
```

```python
#printing the last 10 rows of the dataset
df.tail(10)
```

```python
#obtaining the datatypes(information)
df.info()
```

```python
#statistical analysis of dataset
df.describe()
```

```python
#dropping the blank columns
df.drop(['Status','unnamed1'],axis=1, inplace=True)
#axis =1 refers to entire column
```

```python
#checking the dataset
df
```

```python
#generating total number of null values in the dataset
pd.isnull(df).sum()
```

```python
#dropping null values
df.dropna(inplace=True)
#using inplace=True for saving the result
```

```python
df.shape
```

```python
#changing datatype of a column
df['Amount']=df['Amount'].astype('int')
```

```python
#checking the datatype
df['Amount'].dtypes
```

```python
df.columns
```

```python
#Renaming a column
df.rename(columns={'Cust_name':'Customer_Name'})
```

```python
#using describe function for specific columns
df[['Age','Orders','Amount']].describe()
```

#EDA


###Gender

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

###Age

```python
df.columns
```

```python
ax=sns.countplot(data=df, x='Age Group', hue="Gender")
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

From the above graph, we can see that females aged between 26-35 spend more as compared to males

###State

```python
df.columns
```

```python
#total number of orders from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
#total amount from top 10 states

sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

From the above graph, we can see that most of the orders and total amount spent were given by Uttar Pradesh, Maharashtra and Karnataka respectively

###Marital Status

```python
ax=sns.countplot(data=df, x='Marital_Status')

sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=sales_state)
```

From the above graph, we can say that most of the buyers are married(woman) and have high purchasing power

###Occupation

```python
sns.set(rc={'figure.figsize':(25,5)})

ax=sns.countplot(data=df, x='Occupation',hue='Occupation')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_occ=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',hue='Occupation',data=sales_occ)
```

###Product Category
